```python
def first_non_repeating(s):
    count = {}

    # count frequency
    for ch in s:
        count[ch] = count.get(ch, 0) + 1

    # find first char with count 1
    for ch in s:
        if count[ch] == 1:
            return ch

    return None


s = "aabbcd"
print(first_non_repeating(s))
```

```python
# Number of rows for the diamond (upper half)
rows = 5


# ---------- Upper Half of Diamond ----------
# Loop from 1 to rows (1,2,3,4,5)
for i in range(1, rows + 1):

    # Print spaces first to center the stars
    # Example: if rows=5 and i=1 → 4 spaces
    # This shifts stars to the right
    spaces = " " * (rows - i)

    # Print stars
    # "* " * i means print star i times
    # Example: i=3 → "* * * "
    stars = "* " * i

    # Print spaces + stars
    print(spaces + stars)


# ---------- Lower Half of Diamond ----------
# Loop from rows-1 down to 1 (4,3,2,1)
# We start from rows-1 because the middle row is already printed
for i in range(rows - 1, 0, -1):

    # Print spaces again to keep pattern centered
    spaces = " " * (rows - i)

    # Print stars decreasing
    stars = "* " * i

    # Print spaces + stars
    print(spaces + stars)
```

```python
# SUMMARY: append() vs extend() in Python

# append()
# - Adds ONE element to the end of the list
# - The element can be any object (number, string, list, etc.)
# - If a list is appended, the entire list becomes a single nested element

# Example
a = [1, 2, 3]
a.append([4, 5])
# Result: [1, 2, 3, [4, 5]]


# extend()
# - Adds MULTIPLE elements to the list
# - Takes an iterable (list, tuple, set, string, etc.)
# - Each element is added individually to the list

# Example
b = [1, 2, 3]
b.extend([4, 5])
# Result: [1, 2, 3, 4, 5]


# Key Difference
# append() → adds the whole object as a single element
# extend() → adds each element from the iterable separately


# Quick Trick to Remember
# append = ADD ONE
# extend = EXPAND LIST
```

```python
# Function to flatten the nested list
def flatten_list(nested_list):

    result = []   # This will store the final flattened list

    # Loop through every element in the list
    for item in nested_list:

        # Check if the item is a list
        if isinstance(item, list):

            # If it is a list, call the function again (recursion)
            # and extend the result with the returned values
            result.extend(flatten_list(item))

        else:
            # If it is not a list (i.e., an integer)
            # simply append it to the result list
            result.append(item)

    return result   # Return the flattened result


# Input list
data = [1,2,[3,[4,[6,8],5],0],4]

# Calling the function
output = flatten_list(data)

# Printing the result
print(output)
```

```python
import numpy as np

# ---------------------------------------------------------
# Step 1: Model predicted probabilities for next token
# (Example probabilities from a language model)
# ---------------------------------------------------------

tokens = ["cat", "dog", "tiger", "lion", "elephant", "zebra"]

probs = np.array([0.40, 0.25, 0.15, 0.10, 0.05, 0.05])

print("Original Tokens:", tokens)
print("Original Probabilities:", probs)

# =========================================================
# 1️⃣ TEMPERATURE SAMPLING
# =========================================================

"""
Temperature controls randomness in predictions.

Low temperature (<1):
    - Makes distribution sharper
    - Model becomes more confident
    - Less creativity

High temperature (>1):
    - Flattens distribution
    - More randomness
    - More creative outputs
"""

temperature = 0.7   # change to test behavior

# Convert probabilities to logits
logits = np.log(probs)

# Scale logits using temperature
scaled_logits = logits / temperature

# Convert back to probabilities using softmax
temperature_probs = np.exp(scaled_logits) / np.sum(np.exp(scaled_logits))

print("\n--- Temperature Sampling ---")
print("Temperature:", temperature)
print("New Probabilities:", temperature_probs)

# Randomly choose token using new distribution
temp_choice = np.random.choice(tokens, p=temperature_probs)

print("Selected Token:", temp_choice)


# =========================================================
# 2️⃣ TOP-K SAMPLING
# =========================================================

"""
Top-K sampling keeps only the K highest probability tokens
and removes the rest.

Example:
K = 3
We only keep the top 3 tokens with highest probability.
"""

k = 3

# Get indices of top K probabilities
top_k_indices = np.argsort(probs)[-k:]

# Extract top K probabilities
top_k_probs = probs[top_k_indices]

# Normalize probabilities so they sum to 1
top_k_probs = top_k_probs / np.sum(top_k_probs)

# Corresponding tokens
top_k_tokens = [tokens[i] for i in top_k_indices]

print("\n--- Top-K Sampling ---")
print("K value:", k)
print("Allowed Tokens:", top_k_tokens)
print("Normalized Probabilities:", top_k_probs)

# Randomly select token from top K
topk_choice = np.random.choice(top_k_tokens, p=top_k_probs)

print("Selected Token:", topk_choice)


# =========================================================
# 3️⃣ TOP-P (NUCLEUS SAMPLING)
# =========================================================

"""
Top-P sampling selects the smallest set of tokens whose
cumulative probability >= P.

Example:
P = 0.8
We keep adding tokens until cumulative probability >= 0.8
"""

p = 0.8

# Sort tokens by probability (highest first)
sorted_indices = np.argsort(probs)[::-1]

sorted_probs = probs[sorted_indices]
sorted_tokens = [tokens[i] for i in sorted_indices]

# Compute cumulative probability
cumulative_probs = np.cumsum(sorted_probs)

# Select tokens until cumulative probability exceeds p
selected_indices = cumulative_probs <= p

nucleus_probs = sorted_probs[selected_indices]
nucleus_tokens = [sorted_tokens[i] for i in range(len(sorted_tokens)) if selected_indices[i]]

# Normalize probabilities
nucleus_probs = nucleus_probs / np.sum(nucleus_probs)

print("\n--- Top-P Sampling ---")
print("P value:", p)
print("Allowed Tokens:", nucleus_tokens)
print("Normalized Probabilities:", nucleus_probs)

# Randomly sample token
topp_choice = np.random.choice(nucleus_tokens, p=nucleus_probs)

print("Selected Token:", topp_choice)
```

```python
# Temperature:
# Controls randomness in token probability distribution.

# Top-K:
# Selects next token only from top K highest probability tokens.

# Top-P (Nucleus):
# Selects smallest set of tokens whose cumulative probability >= P.

# Modern LLMs usually use:
# temperature + top_p together.
```

```python
# CACHE vs PERSIST (Apache Spark)

# cache()
# - Used to store intermediate data in memory
# - Shortcut for persist(MEMORY_ONLY)
# - Data is kept only in RAM
# - Faster access but may fail if memory is not enough

# Example
df.cache()


# persist()
# - More flexible version of cache
# - Allows storing data in different storage levels
# - Can store in memory, disk, or both
# - Used when dataset is large

# Example
df.persist()


# Storage levels available in persist
# MEMORY_ONLY      → store only in RAM
# MEMORY_AND_DISK  → store in RAM, if full then use disk
# DISK_ONLY        → store only on disk


# Key Difference
# cache()   → default memory storage (simple)
# persist() → customizable storage levels


# Shortcut to remember
# cache()  = quick memory storage
# persist() = flexible storage (memory + disk options)
```

```python
# SERIALIZATION vs DESERIALIZATION

# Serialization
# - Process of converting an object (data structure) into a format
#   that can be stored or transmitted
# - Usually converted into formats like JSON, bytes, or string
# - Used when sending data over a network or saving to a file

# Example (Serialization)
import json

data = {"name": "Indra", "age": 21}

serialized_data = json.dumps(data)  # convert dictionary → JSON string
print(serialized_data)


# Deserialization
# - Process of converting the stored/transmitted format
#   back into the original object
# - Converts JSON/string/bytes → Python object

# Example (Deserialization)
deserialized_data = json.loads(serialized_data)  # JSON string → dictionary
print(deserialized_data)


# Key Difference
# Serialization   → Object → File / JSON / Bytes
# Deserialization → File / JSON / Bytes → Object


# Real-world usage
# - APIs send data in JSON format
# - Databases store serialized data
# - Distributed systems send serialized objects between machines


# Simple Trick to Remember
# Serialization   = Save / Send the object
# Deserialization = Read / Reconstruct the object
```

```python
def sanity_check(data_type):
  def outer_wrapper(func):
    def inner_wrapper(*args):
      if type(*args) == data_type:
        func(*args)
      else:
        raise TypeError('Ye datatype nai chalega')
    return inner_wrapper
  return outer_wrapper

@sanity_check(int)
def square(num):
  print(num**2)

@sanity_check(str)
def greet(name):
  print('hello',name)

square(2)
```

```python
import time

# Creating data structures
lst = list(range(1000000))
tpl = tuple(range(1000000))
st = set(range(1000000))
dct = {i: i for i in range(1000000)}

# Element to search
target = 999999

# List retrieval
start = time.time()
target in lst
end = time.time()
print("List retrieval time:", end - start)

# Tuple retrieval
start = time.time()
target in tpl
end = time.time()
print("Tuple retrieval time:", end - start)

# Set retrieval
start = time.time()
target in st
end = time.time()
print("Set retrieval time:", end - start)

# Dictionary retrieval
start = time.time()
target in dct
end = time.time()
print("Dictionary retrieval time:", end - start)
```

```python
# Summary: Retrieval Speed Comparison in Python Data Structures

# 1. List
# - Retrieval/Search: O(n)
# - Reason: Elements are stored sequentially.
# - Python must check each element one by one (linear search).
# - Slow when dataset becomes large.

# 2. Tuple
# - Retrieval/Search: O(n)
# - Reason: Similar to list (sequential storage).
# - Slightly faster than list because tuples are immutable.

# 3. Set
# - Retrieval/Search: O(1) average
# - Reason: Uses a hash table internally.
# - Python computes the hash and directly finds the element.
# - Very fast for membership checking (x in set).

# 4. Dictionary
# - Retrieval/Search: O(1) average
# - Reason: Also implemented using a hash table.
# - Uses keys to directly access values.
# - Fastest for key-based lookup.

# Set vs Dictionary Comparison
# - Both use hash tables → O(1) lookup.
# - Set stores only values → slightly less memory.
# - Dictionary stores key-value pairs → slightly more overhead.
# - Dictionary is best when you need mapping (key → value).
# - Set is best when you only need fast membership checking.
```

```python
# Function to find factorial    strong number
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

# Input number
num = int(input("Enter a number: "))

temp = num
sum_fact = 0

# Calculate sum of factorial of digits
while temp > 0:
    digit = temp % 10
    sum_fact += factorial(digit)
    temp //= 10

# Check strong number
if sum_fact == num:
    print(num, "is a Strong Number")
else:
    print(num, "is not a Strong Number")
```

```python
import threading

# function that prints numbers
def print_numbers():
    for i in range(5):
        print(i)

# create first thread
thread1 = threading.Thread(target=print_numbers)

# create second thread
thread2 = threading.Thread(target=print_numbers)

# start both threads
thread1.start()
thread2.start()

# wait until both threads finish
thread1.join()
thread2.join()
```

```python
# GIL (Global Interpreter Lock)
# In CPython only one thread executes Python bytecode at a time
# Prevents memory conflicts and ensures thread safety
# Because of GIL, multithreading does NOT improve CPU-bound tasks

# Multithreading
# Running multiple threads inside a single process
# Threads share the same memory
# Best used for I/O-bound tasks (file reading, API calls, database operations)

# Multiprocessing
# Running multiple processes simultaneously
# Each process has its own memory space
# Not affected by GIL
# Best used for CPU-bound tasks (ML training, heavy calculations, image processing)

# Async Programming
# Uses a single thread with an event loop
# Handles many tasks concurrently without creating multiple threads
# Best for high-concurrency I/O tasks (many API calls, web scraping)

# Quick Comparison
# Multithreading → best for I/O tasks
# Multiprocessing → best for CPU tasks
# Async → best for handling many waiting operations efficiently
```

```python
import asyncio
import nest_asyncio
nest_asyncio.apply()

# async function definition
async def say_hello():
    print("Hello")
    await asyncio.sleep(2)   # pause for 2 seconds without blocking
    print("World")

# run the async function
asyncio.run(say_hello())
```

```python
# async keyword → defines an asynchronous function

# await → pauses the function until the task completes
# but allows other tasks to run during the wait

# asyncio.sleep() → non-blocking sleep (used instead of time.sleep())

# asyncio.run() → starts the event loop and runs the async function
```

```python
import asyncio

async def task1():
    print("Task 1 started")
    await asyncio.sleep(2)
    print("Task 1 finished")

async def task2():
    print("Task 2 started")
    await asyncio.sleep(2)
    print("Task 2 finished")

async def main():
    await asyncio.gather(task1(), task2())

asyncio.run(main())
```

```python
# asyncio is used for asynchronous programming in Python
# allows multiple I/O tasks to run concurrently
# uses event loop instead of multiple threads
# async → defines asynchronous function
# await → waits for a task without blocking program
# asyncio.run() → starts the event loop
# best used for API calls, network operations, and web scraping
```

```python
# ---------------------------------------------------
# 1. Normalize a list of numbers (used in ML preprocessing)
# ---------------------------------------------------

data = [10, 20, 30, 40, 50]

# find minimum and maximum values
min_val = min(data)
max_val = max(data)

# normalize values between 0 and 1
normalized = [(x - min_val) / (max_val - min_val) for x in data]

print("Normalized data:", normalized)


# ---------------------------------------------------
# 2. Mean, Median, Mode (basic statistics for ML)
# ---------------------------------------------------

import statistics

numbers = [1,2,2,3,4,5]

mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)


# ---------------------------------------------------
# 3. Remove Outliers (simple threshold method)
# ---------------------------------------------------

data = [10,12,13,15,100,14,11]

threshold = 50

# keep values less than threshold
filtered_data = [x for x in data if x < threshold]

print("Filtered data:", filtered_data)


# ---------------------------------------------------
# 4. Word Frequency Counter (NLP task)
# ---------------------------------------------------

text = "ai is powerful and ai is the future"

words = text.split()

freq = {}

# count frequency of each word
for word in words:
    freq[word] = freq.get(word,0) + 1

print("Word frequency:", freq)


# ---------------------------------------------------
# 5. Cosine Similarity (basic NLP similarity)
# ---------------------------------------------------

from collections import Counter
import math

text1 = "ai is powerful"
text2 = "ai is future"

words1 = text1.split()
words2 = text2.split()

counter1 = Counter(words1)
counter2 = Counter(words2)

# get all unique words
all_words = set(counter1.keys()).union(set(counter2.keys()))

# calculate dot product
dot_product = sum(counter1[w] * counter2[w] for w in all_words)

# calculate magnitude
mag1 = math.sqrt(sum(counter1[w]**2 for w in all_words))
mag2 = math.sqrt(sum(counter2[w]**2 for w in all_words))

similarity = dot_product / (mag1 * mag2)

print("Cosine similarity:", similarity)


# ---------------------------------------------------
# 6. Train-Test Split (used in ML model training)
# ---------------------------------------------------

import random

data = list(range(100))

# shuffle dataset
random.shuffle(data)

# 80% training data
split = int(0.8 * len(data))

train = data[:split]
test = data[split:]

print("Train size:", len(train))
print("Test size:", len(test))


# ---------------------------------------------------
# 7. Simple Linear Regression Concept
# ---------------------------------------------------

x = [1,2,3,4]
y = [2,4,6,8]

# simple slope calculation concept
m = sum(y) / sum(x)

# predicted values
predicted = [m * i for i in x]

print("Predicted values:", predicted)


# ---------------------------------------------------
# 8. One-Hot Encoding (convert categories to vectors)
# ---------------------------------------------------

categories = ["cat","dog","bird"]

unique = list(set(categories))

# create encoding dictionary
encoding = {cat:[1 if i==cat else 0 for i in unique] for cat in unique}

print("One-hot encoding:", encoding)


# ---------------------------------------------------
# 9. Tokenization (basic NLP preprocessing)
# ---------------------------------------------------

text = "Machine learning is amazing"

# convert to lowercase and split words
tokens = text.lower().split()

print("Tokens:", tokens)


# ---------------------------------------------------
# 10. Top K Frequent Words (NLP task)
# ---------------------------------------------------

from collections import Counter

text = "ai ai ml data ai ml python data"

words = text.split()

# find top 2 frequent words
top_words = Counter(words).most_common(2)

print("Top frequent words:", top_words)
```
